## 자발적 폐지기업은 제외하기 

```python
df = pd.read_csv('./y/상장폐지현황.csv')
```

```python
# 날짜재구성
df["폐지일자"] = df["폐지일자"].map(change_year)
```

```python
#컬럼명 변경
df = df.rename(columns={'폐지일자':'회계년도','종목코드':'거래소코드'})
```

```python
#회사명 스팩인것 제외
df = df[~df['회사명'].str.contains('스팩')]
```

```python
# 일정 문자가 들어간 행 삭제
df = df[~df['폐지사유'].str.contains('코스닥시장 상장')]
df = df[~df['폐지사유'].str.contains('코스닥시장 이전상장')]
df = df[~df['폐지사유'].str.contains('유가증권시장 상장')]
df = df[~df['폐지사유'].str.contains('상장폐지 신청')]
df = df[~df['폐지사유'].str.contains('상장폐지신청')]
df = df[~df['폐지사유'].str.contains('신청에 의한 상장폐지')]
df = df[~df['폐지사유'].str.contains('종목별 상장폐지신청')]
df = df[~df['폐지사유'].str.contains('완전자회사')]
df = df[~df['폐지사유'].str.contains('존속기간 만료')]
df = df[~df['폐지사유'].str.contains('SPAC 상장예비심사청구서 미제출')]
df = df[~df['폐지사유'].str.contains('포괄적 주식교환에 따른 상장폐지 신청')]
```

```python
# 피흡수합병사유 중 과거 관리종목이 된것 제외하고 자발적으로 봄 
df = df[~(df['폐지사유'].str.contains('피흡수합병')) | (df['회사명'].isin(['알덱스','더체인지']))]
```

```python
# 한진해운을 제외하고 해산 사유 발생은 피흡수합병과 동일하다는것을 확인
df = df[(df['폐지사유'] != '해산 사유 발생') | (df['회사명'].isin(['한진해운']))]
```

```python
# 스팩주
df = df[~df['폐지사유'].str.contains('상장예비심사 청구서 미제출로 관리종목 지정 후 1개월 이내 동 사유 미해소')]
```

```python
# 코넥스만 있는거 확인
df = df[~df['폐지사유'].str.contains('지정자문인')]
```
